```python
# # Есеп 1.1: Класс құру
class Kitap:
    def __init__(self, atau, avtory):
        self.atau = atau
        self.avtory = avtory

    def kitap_turaly_malimet(self):
        print(f"Кітап: {self.atau}, Авторы: {self.avtory}")

# Нысан (объект) жасау
meni_kitap = Kitap("Үш мушкетер", "А. Дюма")

# Әдісті шақыру
meni_kitap.kitap_turaly_malimet()
```

```python
import numpy as np

# Бір өлшемді массив
arr_a = np.arange(1, 10)  # 1-ден 9-ға дейін
print("arr_a:", arr_a)
print("Өлшемі:", arr_a.shape)
print("Деректер түрі:", arr_a.dtype)
print()

# Екі өлшемді массив (3x3)
arr_b = np.arange(1, 10).reshape(3, 3)
print("arr_b:\n", arr_b)
print("Өлшемі:", arr_b.shape)
print("Деректер түрі:", arr_b.dtype)
```

```python
import numpy as np

# Кездейсоқ векторлар жасау
vec1 = np.random.randint(1, 11, size=5)  # 1-ден 10-ға дейін
vec2 = np.random.randint(1, 11, size=5)

print("vec1:", vec1)
print("vec2:", vec2)

# Векторлық қосу
sum_vec = vec1 + vec2
print("Қосынды (vec1 + vec2):", sum_vec)

# Әр элементтің квадраты
vec1_sq = vec1 ** 2
print("vec1 элементтерінің квадраты:", vec1_sq)

# Орташа мән
mean_val = vec1.mean()
print("vec1 орташа мәні:", mean_val)

# Ең үлкен элемент
max_val = vec1.max()
print("vec1 ең үлкен мәні:", max_val)
```
