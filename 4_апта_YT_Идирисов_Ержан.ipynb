```python
# 1. Rectangle класы (ауданы мен периметрін есептеу)
class Rectangle:
    def __init__(self, width, height):

# Параметрлер: width - ені, height - биіктігі
        self.width = width
        self.height = height

    def area(self):
# Ауданын есептейтін әдіс
        return self.width * self.height

    def perimeter(self):
# Периметрін есептейтін әдіс
        return 2 * (self.width + self.height)

# Мысал
rect = Rectangle(5, 5)
print("Аудан:", rect.area())
print("Периметр:", rect.perimeter())
```

```python
# 2. Файлды ашу және оқу try-except арқылы
try:
    filename = input("Файл атын енгізіңіз: ")
    with open(filename, 'r', encoding='utf-8') as f:
        content = f.read()
        print("Файл мазмұны:")
        print(content)
except FileNotFoundError:
  # Файл табылмаған жағдайда
    print("Мұндай файл табылмады!")
except PermissionError:
  # Файлға рұқсат жоқ болған жағдайда
    print("Файлды ашуға рұқсат жоқ!")
finally:
  # Әрқашан орындалатын блок
    print("Бағдарлама орындалды.")
```

```python
# 3. Өз модульіңізді құру (mymath.py функциялары)
# 1. Қосу функциясы
def add(a, b):
    """
    Екі санды қосу.
    Параметрлер:
    a, b: Қосылатын сандар
    Қайтарым: Қосынды
    """
    return a + b

# 2. Алу функциясы
def subtract(a, b):
    """
    Екі санның айырымын есептеу.
    """
    return a - b

# 3. Көбейту функциясы
def multiply(a, b):
    """
    Екі санды көбейту.
    """
    return a * b

# 4. Бөлу функциясы
def divide(a, b):
    """
    Екі санды бөлу.
    Егер b = 0 болса, хабарлама шығарады.
    """
    if b == 0:
        return "Нөлге бөлуге болмайды!"
    return a / b

# 5. Модульдің өздігінен тексеру бөлімі
if __name__ == "__main__":
    # Модульді тікелей іске қосқан кезде тексеру
    print("5 + 3 =", add(5, 3))
    print("5 - 3 =", subtract(5, 3))
    print("5 * 3 =", multiply(5, 3))
    print("5 / 3 =", divide(5, 3))
```

```python
#4. NumPy арқылы кездейсоқ массив жасау және статистикасын шығару
import numpy as np

# 10 элементтен тұратын кездейсоқ массив
arr = np.random.randint(1, 101, size= 10)  # 1-100 аралығында
print("Массив:", arr)

# Орта мән, максимум, минимум
print("Орта мән:", np.mean(arr))
print("Максимум:", np.max(arr))
print("Минимум:", np.min(arr))
```
