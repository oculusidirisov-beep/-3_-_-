```python
# students_max_score_ties.py
students = {
    "Аружан": 88,
    "Медет": 92,
    "Ғани": 75,
    "Диас": 92,
    "Лаура": 81
}

max_score = max(students.values())            # ең үлкен санды табамыз
top_students = [name for name, score in students.items() if score == max_score]

print(f"Ең жоғары баға: {max_score}")
print("Оны алған студент(тер):", ", ".join(top_students))
```
