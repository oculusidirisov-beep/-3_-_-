```python
import json

# 1. Файл оқу
with open('numbers.txt', 'r', encoding='utf-8') as f:
    numbers = f.read().split()
    numbers = [int(n) for n in numbers]
    total = sum(numbers)
    print(" Файлдағы сандар қосындысы:", total)


```

```python
# 3. Dictionary
students = {
    "Айжан": 85,
    "Мұрат": 92,
    "Гүлнұр": 95,
    "Ержан": 98
}
max_score = max(students.values())
top_student = [name for name, score in students.items() if score == max_score]
print(f" Ең жоғары баға {max_score}, студент(тер): {top_student}")

```

```python


# 2. List comprehension
cubes = [n**3 for n in range(1, 21) if n % 3 == 0]
print(" 3-ке бөлінетін сандардың кубтары:", cubes)

```

```python
# 4. Set
numbers_list = [1, 2, 2, 3, 4, 4, 5, 5, 5]
unique_numbers = set(numbers_list)
print(" Қайталанбайтын элементтер жиыны:", unique_numbers)
```

```python
# 5. JSON
students_data = [
    {"name": "Айжан", "grade": 85},
    {"name": "Мұрат", "grade": 92},
    {"name": "Гүлнұр", "grade": 78},
    {"name": "Ержан", "grade": 98}
]

# JSON файлға жазу
with open('students.json', 'w', encoding='utf-8') as f:
    json.dump(students_data, f, ensure_ascii=False, )

# Орташа бағаны есептеу
avg = sum(s['grade'] for s in students_data) / len(students_data)
print(f" Орташа баға: {avg:.2f}")
```
