```python
print("Hello World!")
# Тапсырма 1
student_name = "Ержан"
student_gpa = 3.5
student_course = 2
is_special = True

print("Айнымалылардың типтері:")
print(type(student_name))
print(type(student_gpa))
print(type(student_course))
print(type(is_special))

# Тапсырма 2
if student_gpa >= 3.5:
    base_grant = 50000
elif 3.0 <= student_gpa < 3.5:
    base_grant = 25000
else:
    base_grant = 0

print("Базалық грант мөлшері:", base_grant, "теңге")

# Тапсырма 3
bonuses = ["math", "physics", "history", "informatics", "english"]

print("\nОлимпиадалар тізімі:")
for b in bonuses:
    print("-", b)

if len(bonuses) >= 3:
    base_grant += 5000

if base_grant == 0:
    i = 0
    while i < 3:
        print("Грант алу үшін GPA-ды көтеру керек!")
        i += 1

print("Олимпиададан кейінгі грант:", base_grant, "теңге")

# Тапсырма 4
def social_bonus(special_status):
    if special_status:
        return 10000
    else:
        return 0

extra_bonus = social_bonus(is_special)
print("Әлеуметтік жеңілдік:", extra_bonus, "теңге")

# Тапсырма 5
total_grant = base_grant + extra_bonus
print(f"\nСтудент {student_name} үшін жалпы грант мөлшері: {total_grant} теңге.")
```
