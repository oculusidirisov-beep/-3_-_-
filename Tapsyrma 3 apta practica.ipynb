```python
import csv
from pprint import pprint

data_rows = []

with open("Salaries.csv", "r", encoding="utf-8") as f:
    reader = csv.reader(f)
    header = next(reader)  # Тақырып жолы
    for i, row in enumerate(reader):
        if i >= 10:
            break
        data_rows.append(row)

print("data_rows ұзындығы:", len(data_rows))
print("Бірінші элемент:")
pprint(data_rows[0])
```

```python
#Тапсырма 2: Кортеж және сөздік (tuple және dict)

#Міндет:

#Алғашқы 3 қызметкердің EmployeeName мәндерін first_three_names кортежіне сақтау.

#Әр қызметкер үшін EmployeeName кілт, JobTitle мән болатын employee_titles #сөздігін жасау.
name_idx = header.index("EmployeeName")
job_idx = header.index("JobTitle")

first_three_names = tuple(row[name_idx] for row in data_rows[:3])

employee_titles = {}
for row in data_rows[:3]:
    employee_titles[row[name_idx]] = row[job_idx]

print("Кортеж (first_three_names):", first_three_names)
print("\nСөздік (employee_titles):")
pprint(employee_titles)

```

```python
#Тапсырма 3: Тізімдік өрнектер (List Comprehension)

#Міндет:
#Барлық 10 қызметкердің BasePay мәндерін мәтін түрінде алу және base_pays_str #тізіміне сақтау.
base_idx = header.index("BasePay")

base_pays_str = [row[base_idx] for row in data_rows]

print("base_pays_str:")
pprint(base_pays_str)
```

```python
#Тапсырма 4: Жиын (Set) — Бірегей атақтарды табу

#Міндет:
#10 қызметкердің JobTitle мәндерін тізімге жинап, содан кейін жиынға (set) #түрлендіру.
job_titles_list = []
for row in data_rows:
    job_titles_list.append(row[job_idx])

unique_titles = set(job_titles_list)

print("Барлық атақтар:")
pprint(job_titles_list)
print("\nБірегей атақтар (unique_titles):")
pprint(unique_titles)

print("\nНеліктен жиын пайдалы?")
print("Жиын қайталанатын мәндерді автоматты түрде алып тастайды,")
print("және бірегей элементтерді анықтау мен іздеу жылдам орындалады.")
```
